# Class 1 overview

Welcome to data analysis and programming. This course uses **Python** as the main programming language, but the larger goal is not only to "learn Python." The goal is to learn a way of thinking and working with data:

1. Ask a clear question.
2. Find or collect relevant data.
3. Clean and organize the data.
4. Analyze patterns.
5. Communicate results honestly and clearly.
6. Make the work reproducible so that you, your classmates, or future collaborators can understand what happened.

Students in this class may come from journalism, social science, humanities, business, natural science, engineering, public health, or other fields. Programming is useful across all of these areas because it helps us handle repetitive tasks, large datasets, transparent analysis, and reproducible writing.

## Learning objectives

By the end of this first class, students should be able to explain:

- What programming is and why Python is useful for data analysis.
- What an **integrated development environment** (IDE) is.
- What **Visual Studio Code** is and how it helps with coding and writing.
- The difference between **Python**, **VS Code**, and the **Python extension for VS Code**.
- What **Git** and **GitHub** are.
- Why plain-text documents such as **Markdown** and **Quarto `.qmd` files** are useful for data analysis, writing, and version control.
- How Markdown/Quarto differs from traditional word processors such as Microsoft Word.

:::

- Combine multiple files.
- Count, group, filter, or summarize records.
- Create charts.
- Run statistical models.
- Generate reports.
- Repeat the same analysis when new data arrives.

# What is Git?

**Git** is a version control system. It records changes to files over time.

Git helps answer questions such as:

- What changed since yesterday?
- Who changed this file?
- Can I go back to an earlier version?
- Can two people work on the same project without emailing files back and forth?
- Can I show the history of my analysis?

## Key Git vocabulary

| Term | Meaning |
|---|---|
| Repository / repo | A project folder tracked by Git |
| Commit | A saved snapshot of changes |
| Commit message | A short explanation of what changed |
| Branch | A separate line of work |
| Merge | Combining changes from different branches |
| Clone | Copying a repository from a remote location to your computer |
| Pull | Bringing remote changes to your computer |
| Push | Sending local commits to a remote repository |
| `.gitignore` | A file that tells Git which files not to track |

## A simple Git workflow

```bash
git status
git add analysis.qmd
git commit -m "Add first analysis draft"
git push
```

In VS Code, many Git actions can also be done through the Source Control panel without typing every command manually.

# What is GitHub?

**GitHub** is a web platform for hosting Git repositories. Git is the version control tool; GitHub is a website that makes Git-based collaboration easier.

GitHub can be used to:

- Store a project online.
- Share code and documents.
- Submit assignments.
- Collaborate with classmates.
- Review changes.
- Track issues and tasks.
- Publish project websites.

::: {.callout-note}
## Git vs GitHub

Git and GitHub are related, but they are not the same thing.

- **Git** is software installed on your computer.
- **GitHub** is an online service that hosts Git repositories.
:::

# Why writing matters in a data analysis course

Data analysis is not complete when the code runs. Students also need to explain:

- What question they asked.
- Where the data came from.
- What assumptions they made.
- What methods they used.
- What results they found.
- What limitations remain.
- What readers should conclude.

For students from journalism and other writing-intensive fields, this is especially important. A good data project combines evidence, computation, and clear writing.

# What is Markdown?

**Markdown** is a lightweight plain-text format for writing structured documents. A Markdown file can be opened in any text editor, and the file remains readable even before it is converted to HTML, PDF, or Word.

Common Markdown syntax:

```markdown
# Heading 1

## Heading 2

This is a paragraph with **bold text** and *italic text*.

- Item one
- Item two
- Item three

[Link text](https://example.com)

| Column A | Column B |
|---|---|
| Value 1 | Value 2 |
```

Markdown is especially useful because it stores writing as plain text. Plain text works well with Git, because Git can clearly show exactly which lines changed.

# What is Quarto?

**Quarto** is an open-source publishing system for technical and scientific documents. A Quarto file usually has the extension `.qmd`.

A `.qmd` file can include:

- Markdown text.
- Python code blocks.
- Tables.
- Figures.
- Citations.
- Equations.
- Callout boxes.
- Links.
- Document metadata.

Quarto can render one source file into multiple output formats, such as HTML, PDF, Word, slides, websites, books, and dashboards.

## A minimal Quarto document

````markdown
---
title: "My First Quarto Document"
format: html
jupyter: python3
---

# Introduction

This document combines writing and code.

`````python
print("Hello, Quarto!")
`````

````

# Why Markdown and Quarto instead of only Microsoft Word?

Microsoft Word is useful and familiar. It is often the right tool for polished documents, tracked comments, administrative forms, and situations where collaborators expect `.docx` files.

However, for data analysis and programming, Markdown and Quarto have several advantages.

| Need | Microsoft Word | Markdown / Quarto |
|---|---|---|
| Human-readable source file | `.docx` is a complex document format | Plain text; readable in any editor |
| Version control with Git | Difficult to see exact line-by-line changes | Excellent; Git can show precise text/code changes |
| Combining code and writing | Usually requires copying results manually | Code, output, and explanation can live in one document |
| Reproducibility | Charts/tables may be pasted manually | Results can be regenerated from code |
| Collaboration through GitHub | Possible, but not natural | Designed for text-based collaboration |
| Multiple output formats | Mainly Word/PDF depending on workflow | Same source can render to HTML, PDF, Word, slides, websites, etc. |
| Learning curve | Familiar interface | Requires learning simple syntax and tools |
| Visual layout control | Strong visual page layout tools | Better for structured writing than fine-grained visual design |

## Example: why version control is easier with plain text

Imagine a student changes this sentence:

```markdown
The city budget increased by 4% in 2024.
```

Later, the sentence becomes:

```markdown
The city budget increased by 6% in 2024 after revised revenue estimates.
```

Git can show the exact line that changed. In a Word document, version history exists, but the internal file format is not as simple for Git to compare, merge, and review.

## When should students use Word?

Word is still useful when:

- The instructor, editor, or organization requires `.docx`.
- A document needs detailed visual layout.
- A collaborator is not comfortable with Markdown, Quarto, or Git.
- The task is mainly final formatting rather than reproducible analysis.

## When should students use Quarto?

Quarto is especially useful when:

- A document includes code and results.
- The analysis may need to be repeated.
- The project benefits from Git version history.
- The writer wants to produce HTML, PDF, Word, or slides from one source file.
- The document should clearly connect evidence, method, and conclusion.

::: {.callout-tip}
## Teaching message

For this course, students do not need to reject Word. Instead, they should learn that different tools serve different purposes. Word is strong for conventional document editing. Markdown and Quarto are strong for transparent, reproducible, version-controlled analytical writing.
:::

# Common beginner questions

## Is VS Code the same as Visual Studio?

No. **Visual Studio Code** is a lightweight editor used for many languages and workflows. **Visual Studio** is a larger development environment often used for software engineering in languages such as C# and C++. This course uses VS Code.

## Is GitHub the same as Google Drive or Dropbox?

No. GitHub stores project history using Git. It is not just file storage. It tracks changes, supports collaboration, and lets people review project history.

## Can I use Word and Quarto together?

Yes. Quarto can render documents to Word format. A common workflow is to write a reproducible analysis in Quarto, then export to Word when a professor, editor, or organization requires `.docx`.

## Do I need to memorize terminal commands?

No. Students should learn a small set of common commands, but the main goal is understanding the workflow. VS Code also provides graphical tools for many tasks.

## Why not use only Jupyter notebooks?

Jupyter notebooks are useful for exploration and interactive coding. Quarto is useful for turning analysis into polished documents, reports, websites, or slides. In this course, we may use both, but `.qmd` files are especially helpful for combining writing, code, and version control.

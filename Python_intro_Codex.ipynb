# 1. What Is Programming?

- Powerful because computers can repeat tasks quickly and consistently.
- Frustrating because small mistakes in instructions can cause errors or wrong results.

# 6. Conditions: Making Decisions

Programs become useful when they can make decisions.

````python
score = 85

if score >= 60:
    print("Pass")
else:
    print("Fail")
````

The `if` statement checks a condition. If the condition is true, Python runs the indented code below it.

## Comparison Operators

| Operator | Meaning |

| `<` | Less than |
| `<=` | Less than or equal to |

````python
age = 20

print(age == 20)
print(age >= 18)
print(age < 18)
````

## `if`, `elif`, And `else`

Use `elif` when there are multiple possible conditions.

````python
score = 87

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(grade)
````

## Boolean Logic

Use `and`, `or`, and `not` to combine conditions.

````python
age = 21
has_id = True

if age >= 21 and has_id:
    print("Entry allowed")
else:
    print("Entry denied")
````

````python
is_weekend = False
is_holiday = True

if is_weekend or is_holiday:
    print("No class today")
else:
    print("Class today")
````

````python
is_raining = False

if not is_raining:
    print("Go for a walk")
````

# 7. Loops: Repeating Work

A loop repeats code.

There are two main loop types in Python:

- `for` loops: repeat over a sequence of items.
- `while` loops: repeat while a condition is true.

## `for` Loops

````python
for number in [1, 2, 3, 4, 5]:
    print(number)
````

The variable `number` takes each value from the list one at a time.

## `range()`

The `range()` function creates a sequence of numbers.

````python
for i in range(5):
    print(i)
````

This prints `0` through `4`. In Python, counting often starts at zero.

````python
for i in range(1, 6):
    print(i)
````

This prints `1` through `5`.

````python
for i in range(0, 10, 2):
    print(i)
````

This prints every second number from `0` up to, but not including, `10`.

## Looping Over Strings

````python
word = "Python"

for letter in word:
    print(letter)
````

## Looping Over Lists

````python
scores = [88, 92, 79, 95]

for score in scores:
    print(score)
````

## `while` Loops

A `while` loop repeats while a condition is true.

````python
count = 1

while count <= 5:
    print(count)
    count += 1
````

Be careful: if the condition never becomes false, the loop runs forever.

````python
# Do not run this unless you know how to stop it.
# count = 1
# while count <= 5:
#     print(count)
````

The problem is that `count` never changes.

## `break` And `continue`

Use `break` to stop a loop early.

````python
for number in range(1, 10):
    if number == 5:
        break
    print(number)
````

Use `continue` to skip to the next loop iteration.

````python
for number in range(1, 6):
    if number == 3:
        continue
    print(number)
````

## Loop Example: Average Score

````python
scores = [88, 92, 79, 95]

total = 0

for score in scores:
    total += score

average = total / len(scores)
print(average)
````

This example introduces `len()`, which returns the number of items in a collection.

A function is a named block of reusable code.

- `print()`

- `int()`

- `str()`

You can also define your own functions.

## Defining A Function

````python
def greet():
    print("Hello!")

greet()
````

## Parameters

Parameters are inputs to a function.

A parameter is a variable name used in a function definition. It tells the function what kind of input it expects.

An argument is the actual value you pass to the function when you call it.

Example:


````python
def greet(name):
    print(f"Hello, {name}!")

greet("Aisha")
greet("Ben")
````

## Return Values

A function can return a result.

````python
def add(a, b):
    return a + b

result = add(3, 4)
print(result)
````

`print()` displays something. `return` sends a value back to the place where the function was called.

## Function Example: Letter Grade

````python
def letter_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

print(letter_grade(95))
print(letter_grade(72))
````

## Why Functions Matter

Functions help us:

- Avoid repeating code.
- Give names to ideas.
- Break large problems into smaller pieces.
- Test one part of a program at a time.

Slicing extracts part of a list.

````python
scores = [88, 92, 79, 95, 84]

print(scores[0:3])
print(scores[2:])
print(scores[:2])
````

## List Comprehensions

A list comprehension is a compact way to create a new list.

````python
numbers = [1, 2, 3, 4, 5]
squares = []

for number in numbers:
    squares.append(number ** 2)

print(squares)
````

The same idea with a list comprehension:

````python
numbers = [1, 2, 3, 4, 5]
squares = [number ** 2 for number in numbers]

print(squares)
````

Filtering with a list comprehension:

````python
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = [number for number in numbers if number % 2 == 0]

print(even_numbers)
````

# 12. Errors And Debugging

Errors are normal. Debugging is the process of finding and fixing problems in code.

## Syntax Errors

A syntax error means Python cannot understand the code structure.

````python
# Missing closing parenthesis
# print("Hello"
````

## Runtime Errors

A runtime error happens while the program is running.

````python
# Division by zero
# print(10 / 0)
````

## Logic Errors

A logic error means the program runs, but the result is wrong.

````python
scores = [80, 90, 100]

# Incorrect average because the denominator should be len(scores)
average = sum(scores) / 2
print(average)
````

Logic errors can be the hardest because Python may not show an error message.

## A Debugging Checklist

When code fails:

1. Read the error message from bottom to top.
2. Find the file name and line number.
3. Check spelling, punctuation, and indentation.
4. Print intermediate values.
5. Test one small part at a time.
6. Ask an AI tool to explain the error, but verify the answer.

Example:

````python
numbers = [10, 20, 30]

for number in numbers:
    print("Current number:", number)
    print("Doubled:", number * 2)
````

## A Tiny Dataset In Pure Python

````python
students = [
    {"name": "Ana", "major": "Biology", "score": 88},
    {"name": "Bo", "major": "History", "score": 92},
    {"name": "Chen", "major": "Computer Science", "score": 79},
    {"name": "Dina", "major": "Economics", "score": 95},
    {"name": "Eli", "major": "Computer Science", "score": 84}
]
````

Question: What is the average score?

````python
total = 0

for student in students:
    total += student["score"]

average = total / len(students)
print(average)
````

Question: How many students are in each major?

````python
major_counts = {}

for student in students:
    major = student["major"]

    if major in major_counts:
        major_counts[major] += 1
    else:
        major_counts[major] = 1

print(major_counts)
````

# 16. Plotting And Visualization

Visualization helps us see patterns that are hard to understand from tables alone.

## Basic Plot With `matplotlib`

````python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5]
temperatures = [70, 72, 71, 75, 78]

plt.plot(days, temperatures)
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Temperature Over Time")
plt.show()
````

## Bar Chart

````python
import matplotlib.pyplot as plt

majors = ["Biology", "History", "Computer Science", "Economics"]
counts = [1, 1, 2, 1]

plt.bar(majors, counts)
plt.xlabel("Major")
plt.ylabel("Number of Students")
plt.title("Students By Major")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
````

## Scatter Plot

````python
import matplotlib.pyplot as plt

hours_studied = [2, 3, 4, 5, 6, 7]
exam_scores = [70, 75, 78, 85, 88, 93]

plt.scatter(hours_studied, exam_scores)
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Study Time And Exam Score")
plt.show()
````

- Near `0`: little linear relationship.
- Near `-1`: strong negative relationship.

# 18. Object-Oriented Programming Preview

Python supports object-oriented programming. This means organizing code around objects that combine data and behavior.

Students do not need to master this immediately, but they should recognize the idea.

## Objects And Methods

In Python, many things are objects.

````python
name = "python"
print(name.upper())
````

Here, `name` is a string object, and `.upper()` is a method.

## Classes

A class is a blueprint for creating objects.

````python
class Student:
    def __init__(self, name, major):
        self.name = name
        self.major = major

    def introduce(self):
        print(f"My name is {self.name} and I study {self.major}.")

student1 = Student("Maya", "Economics")
student1.introduce()
````

In this example:

- `Student` is a class.
- `student1` is an object.
- `name` and `major` are attributes.
- `introduce()` is a method.

## Python, JavaScript, And C# Classes

Python:

````python
class Student:
    def __init__(self, name):
        self.name = name
````

JavaScript:

```javascript
class Student {
  constructor(name) {
    this.name = name;
  }
}
```

C#:

```csharp
class Student
{
    public string Name;

    public Student(string name)
    {
        Name = name;
    }
}
```

C# uses explicit types and access modifiers such as `public`. Python is less formal, which makes small examples shorter.

# 19. Programming In The Age Of AI

AI tools can make programming more accessible, but they do not remove the need to understand code.

## Useful Ways To Use AI

AI can help with:

- Explaining unfamiliar code.
- Explaining error messages.
- Generating small examples.
- Suggesting test cases.
- Translating code between languages.
- Improving comments and documentation.
- Brainstorming approaches to a problem.

## Risky Ways To Use AI

Be careful when AI:

- Produces code you cannot explain.
- Makes up library functions or package names.
- Ignores edge cases.
- Gives a correct-looking but wrong answer.
- Uses outdated syntax.
- Solves an assignment in a way that bypasses learning.

## A Good AI Prompt For Learning

Weak prompt:

```text
Write my Python homework.
```

Better prompt:

```text
I am learning Python loops. I need to count how many numbers in a list are greater than 10. Please explain the idea first, then show a small example, then give me a practice problem without the answer.
```

## A Good AI Prompt For Debugging

```text
I am a beginner in Python. This code gives an error. Please explain what the error means, point to the likely line, and suggest the smallest fix. Do not rewrite the whole program unless necessary.

<paste code here>
<paste full error message here>
```

## Always Verify AI-Generated Code

When AI gives you code:

1. Read it line by line.
2. Run it on a small example.
3. Check whether the output makes sense.
4. Try an edge case.
5. Explain the code in your own words.

# Appendix B: Common Beginner Mistakes

## Confusing `=` And `==`

````python
x = 5      # assignment
print(x == 5)  # comparison
````

## Forgetting Indentation

````python
if x > 0:
    print("Positive")
````

## Mixing Strings And Numbers

````python
age_text = "20"
age = int(age_text)
print(age + 1)
````

## Off-By-One Errors

````python
for i in range(1, 6):
    print(i)
````

This prints `1` through `5`, not `1` through `6`.

## Changing A List While Looping Over It

This can cause confusing behavior:

````python
# Avoid this pattern as a beginner.
# numbers = [1, 2, 3, 4]
# for number in numbers:
#     if number % 2 == 0:
#         numbers.remove(number)
````

Use a new list instead:

````python
numbers = [1, 2, 3, 4]
odd_numbers = []

for number in numbers:
    if number % 2 != 0:
        odd_numbers.append(number)

print(odd_numbers)
````
